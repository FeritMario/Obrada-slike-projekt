```python


import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
import pytesseract

# Učitavanje slike
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
print(f"Prenesena datoteka: {file_name}")
image = cv2.imread(file_name)

# Provjera je li slika ispravno učitana
if image is None:
    raise ValueError("Nije moguće učitati sliku. Provjeri put do slike.")

# Prikaz originalne slike
plt.figure(figsize=(10, 5))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Originalna slika')
plt.axis('off')
plt.show()

# Konvertiranje slike u sivi ton
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Thresholding za binarizaciju slike
_, binary_image = cv2.threshold(gray_image, 200, 255, cv2.THRESH_BINARY_INV)

# Pronalaženje kontura
contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Kopiranje slike za crtanje kontura za vizualizaciju
contour_image = cv2.cvtColor(gray_image, cv2.COLOR_GRAY2BGR)
cv2.drawContours(contour_image, contours, -1, (0, 0, 255), 2)  # Crvena boja za konture (0, 0, 255)

# Prikaz slike s konturama
plt.figure(figsize=(10, 5))
plt.imshow(cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB))
plt.title('Konture na slici')
plt.axis('off')
plt.show()

# Pronalaženje pravokutnika koji okružuje graf
x, y, w, h = cv2.boundingRect(contours[0])

# Ručno definiranje područja interesa na osnovi vizualne procjene
roi_x_axis_precise = gray_image[y + h + 10:y + h + 40, x:x + w]
roi_y_axis_precise = gray_image[y:y + h, x - 35:x - 5]

# Funkcija za podešavanje kontrasta i svjetline
def adjust_contrast_and_brightness(image, contrast=1.5, brightness=0):
    return cv2.convertScaleAbs(image, alpha=contrast, beta=brightness)

# Poboljšanje kontrasta i svjetline za precizno izrezana područja
roi_x_axis_precise = adjust_contrast_and_brightness(roi_x_axis_precise)
roi_y_axis_precise = adjust_contrast_and_brightness(roi_y_axis_precise)

# Ponovna binarizacija precizno izrezanih područja
_, roi_x_axis_precise = cv2.threshold(roi_x_axis_precise, 200, 255, cv2.THRESH_BINARY)
_, roi_y_axis_precise = cv2.threshold(roi_y_axis_precise, 200, 255, cv2.THRESH_BINARY)

# Primjena morfoloških operacija za uklanjanje šuma
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (1, 1))
roi_x_axis_precise = cv2.morphologyEx(roi_x_axis_precise, cv2.MORPH_CLOSE, kernel)
roi_y_axis_precise = cv2.morphologyEx(roi_y_axis_precise, cv2.MORPH_CLOSE, kernel)

# Primjena OCR-a s konfiguracijom za prepoznavanje samo brojeva
custom_config = r'--oem 3 --psm 6 outputbase digits'

text_x_axis_precise = pytesseract.image_to_string(roi_x_axis_precise, config=custom_config)
text_y_axis_precise = pytesseract.image_to_string(roi_y_axis_precise, config=custom_config)

print("Brojevi na X osi:", text_x_axis_precise)
print("Brojevi na Y osi:", text_y_axis_precise)

# Pronalazak lokacije nule
zero_x = x  # X koordinata nule je početak x pravokutnika
zero_y = y + h  # Y koordinata nule je kraj y pravokutnika

# Prikaz slike s označenom nulom
image_with_zero = contour_image.copy()
cv2.circle(image_with_zero, (zero_x, zero_y), 5, (255, 0, 0), -1)  # Plava boja za označavanje nule

plt.figure(figsize=(10, 5))
plt.imshow(cv2.cvtColor(image_with_zero, cv2.COLOR_BGR2RGB))
plt.title('Lokacija nule na slici')
plt.axis('off')
plt.show()

print("Lokacija nule grafa: (", zero_x, ",", zero_y, ")")

# Novi zadatak: Pronalazak grafa na temelju debljine crte ili boje
# Konvertiranje slike u HSV za lakše razlikovanje boja
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Definiranje raspona boja koje odgovaraju crti grafa (npr. crna boja)
lower_black = np.array([0, 0, 0])
upper_black = np.array([180, 255, 30])

# Kreiranje maske za crnu boju
mask = cv2.inRange(hsv_image, lower_black, upper_black)

# Primjena maske na originalnu sliku
result = cv2.bitwise_and(image, image, mask=mask)

# Konvertiranje rezultirajuće slike u sivi ton
gray_result = cv2.cvtColor(result, cv2.COLOR_BGR2GRAY)

# Thresholding za binarizaciju rezultirajuće slike
_, binary_result = cv2.threshold(gray_result, 1, 255, cv2.THRESH_BINARY)

# Pronalaženje kontura na rezultirajućoj slici
contours_graph, _ = cv2.findContours(binary_result, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Kopiranje slike za crtanje kontura grafa za vizualizaciju
contour_image_graph = image.copy()
cv2.drawContours(contour_image_graph, contours_graph, -1, (255, 0, 0), 2)  # Plava boja za konture grafa (255, 0, 0)

# Prikaz slike s konturama grafa
plt.figure(figsize=(10, 5))
plt.imshow(cv2.cvtColor(contour_image_graph, cv2.COLOR_BGR2RGB))
plt.title('Konture grafa na slici')
plt.axis('off')
plt.show()

# Novi zadatak: Pronaći udaljenost grafa od x-osi po stupcima slike

# Pretpostavljamo da je x-osa na dnu bounding box-a grafa
x_axis_y = y + h

# Pronalazak udaljenosti grafa od x-osi po stupcima
height, width = binary_result.shape
distances = []

for col in range(x, x + w):
    # Pronađi sve ne-nula piksele u stupcu
    non_zero_pixels = np.where(binary_result[:, col] != 0)[0]
    if len(non_zero_pixels) > 0:
        # Udaljenost od x-osi do prvog ne-nula piksela u stupcu
        distance = x_axis_y - non_zero_pixels[0]
        distances.append(distance)
    else:
        distances.append(0)

# Prikaz udaljenosti
plt.figure(figsize=(10, 5))
plt.plot(range(x, x + w), distances)
plt.title('Udaljenost grafa od x-osi po stupcima')
plt.xlabel('Stupac (x)')
plt.ylabel('Udaljenost od x-osi (piksela)')
plt.show()

# Pretvaranje teksta s oznakama na osi u brojeve
x_values = [int(val) for val in text_x_axis_precise.split() if val.isdigit()]
y_values = [int(val) for val in text_y_axis_precise.split() if val.isdigit()]

# Nacrtati graf s vrijednostima x i y
plt.figure(figsize=(10, 5))
plt.plot(x_values, y_values, marker='o', linestyle='-')
plt.title('Graf s oznakama na osima')
plt.xlabel('X vrijednosti')
plt.ylabel('Y vrijednosti')

# Usporediti s originalnom slikom
plt.figure(figsize=(10, 5))

# Originalna slika
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Originalna slika')
plt.axis('off')

# Nacrtati graf s oznakama na osima
plt.subplot(1, 2, 2)
plt.plot(x_values, y_values, marker='o', linestyle='-')
plt.title('Graf s oznakama na osima')
plt.xlabel('X vrijednosti')
plt.ylabel('Y vrijednosti')

# Prikazati
plt.tight_layout()
plt.show()

```
